```python
from rdkit import Chem
import numpy as np
from __future__ import division
%matplotlib inline
import scipy as sp
import matplotlib.pyplot as plt
import matplotlib as mpl
from rdkit.Chem import rdMolDescriptors
```

```python
with open('smiles_files/10000_smiles.txt') as f:
    original = f.readlines()
original = [i.strip('\n') for i in original]
with open('smiles_files/25000_random_smiles.txt') as f:
    small = f.readlines()
small = [i.strip('\n') for i in small]
with open('smiles_files/large_smiles') as f:
    large = f.readlines()
large = [i.strip('\n') for i in large]
with open('smiles_files/random_smiles.txt') as f:
    random = f.readlines()
random = [i.strip('\n') for i in random]
```

```python
mol_weights = []
atom_nums = []
bond_nums = []
ring_num = []
for i in original:
    mol = Chem.MolFromSmiles(i)
    mol_weights.append(rdMolDescriptors.CalcExactMolWt(mol))
    atom_nums.append(mol.GetNumAtoms())
    bond_nums.append(mol.GetNumBonds())
    ring_num.append(rdMolDescriptors.CalcNumRings(mol))
```

```python
fig= plt.figure(figsize=(16, 8))
ax1 = fig.add_subplot(2,2,1)
ax1.set_xlabel("Molecular Weight (g/mol)")
ax1.set_ylabel("Count")
n_1, bins_1, patches_1 = ax1.hist(mol_weights,100)
ax2 = fig.add_subplot(2,2,2)
ax2.set_xlabel("Number of Atoms")
ax2.set_ylabel("Count")
n_2, bins_2, patches_2 = ax2.hist(atom_nums,100)
ax3 = fig.add_subplot(2,2,3)
ax3.set_xlabel("Number of Bonds")
ax3.set_ylabel("Count")
n_3, bins_3, patches_3 = ax3.hist(bond_nums,100)
ax4 = fig.add_subplot(2,2,4)
ax4.set_xlabel("Number of Rings")
ax4.set_ylabel("Count")
n_4, bins_4, patches_4 = ax4.hist(ring_num, 100)
```
