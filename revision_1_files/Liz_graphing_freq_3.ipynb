```python
%matplotlib inline
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from collections import defaultdict
from __future__ import division
```

```python
### Chemical Data
chem_data = defaultdict(dict)
with open('data/1000_out_2.txt') as f:
    for i in f.readlines():
        k,v = i.strip('\n').split('\t')
        chem_data['Random_1000'][v] = int(k)
with open('data/small_dist_1000_out.txt') as f:
    for i in f.readlines():
        k,v = i.strip('\n').split('\t')
        chem_data['Small_1000'][v] = int(k)
```

```python
final = defaultdict(dict)

i = "Random_1000"
tmp = chem_data[i]
test = sorted(tmp.values(), reverse = True)
t = sum(test)
test = [j/t for j in test]
final[i]['chem_data'] = test
final[i]['color'] = 'Green'

i = "Small_1000"
tmp = chem_data[i]
test = sorted(tmp.values(), reverse = True)
t = sum(test)
test = [j/t for j in test]
final[i]['chem_data'] = test
final[i]['color'] = 'Blue'
```
