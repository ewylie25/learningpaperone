```python
%matplotlib inline
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from collections import defaultdict
from __future__ import division
```

```python
### Chemical Data
chem_data = defaultdict(dict)
with open('data/smarts_stats_30.tsv') as f:
    for i in f.readlines():
        k,v = i.strip('\n').split('\t')
        chem_data['Original_30'][v] = int(k)
with open('data/random_30.txt') as f:
    for i in f.readlines():
        k,v = i.strip('\n').split('\t')
        chem_data['Random_30'][v] = int(k)
with open('data/random_30_2.txt') as f:
    for i in f.readlines():
        k,v = i.strip('\n').split('\t')
        chem_data['Random_30_2'][v] = int(k)
with open('data/random_60.txt') as f:
    for i in f.readlines():
        k,v = i.strip('\n').split('\t')
        chem_data['Random_60'][v] = int(k)
with open('data/random_60_2.txt') as f:
    for i in f.readlines():
        k,v = i.strip('\n').split('\t')
        chem_data['Random_60_2'][v] = int(k)
with open('data/smarts_stats_60.tsv') as f:
    for i in f.readlines():
        k,v = i.strip('\n').split('\t')
        chem_data['Original_60'][v] = int(k)
with open('data/smarts_stats_1000.tsv') as f:
    for i in f.readlines():
        k,v = i.strip('\n').split('\t')
        chem_data['Original_1000'][v] = int(k)
with open('data/redo_orig_1000.txt') as f:
    for i in f.readlines():
        k,v = i.strip('\n').split('\t')
        chem_data['Redo_Original_1000'][v] = int(k)


### English data
eng_data = defaultdict(dict)
with open('data/en_stats_30.tsv') as f:
    for i in f.readlines():
        k,v =i.strip('\n').split('\t')
        eng_data['30'][v] = int(k)
with open('data/en_stats_60.tsv') as f:
    for i in f.readlines():
        k,v =i.strip('\n').split('\t')
        eng_data['60'][v] = int(k)
```

```python
final = defaultdict(dict)

i = "Original_30"
tmp = chem_data[i]
test = sorted(tmp.values(), reverse = True)
t = sum(test)
test = [j/t for j in test]
final[i]['chem_data'] = test
final[i]['color'] = 'Green'

i = "Random_30"
tmp = chem_data[i]
test = sorted(tmp.values(), reverse = True)
t = sum(test)
test = [j/t for j in test]
final[i]['chem_data'] = test
final[i]['color'] = 'Blue'

tmp = chem_data["Random_30_2"]
test = sorted(tmp.values(), reverse = True)
t = sum(test)
test = [j/t for j in test]
final["Random_30_2"]['chem_data'] = test

final["Random_30_2"]['color'] = 'Teal'

i = "Original_60"
tmp = chem_data[i]
test = sorted(tmp.values(), reverse = True)
t = sum(test)
test = [j/t for j in test]
final[i]['chem_data'] = test
final[i]['color'] = 'Orange'

tmp = chem_data["Random_60"]
test = sorted(tmp.values(), reverse = True)
t = sum(test)
test = [j/t for j in test]
final["Random_60"]['chem_data'] = test
final["Random_60"]['color'] = 'Red'

tmp = chem_data["Random_60_2"]
test = sorted(tmp.values(), reverse = True)
t = sum(test)
test = [j/t for j in test]
final["Random_60_2"]['chem_data'] = test
final["Random_60_2"]['color'] = 'Purple'

tmp = chem_data["Original_1000"]
test = sorted(tmp.values(), reverse = True)
t = sum(test)
test = [j/t for j in test]
final["Original_1000"]['chem_data'] = test
final["Original_1000"]['color'] = 'Magenta'

tmp = chem_data["Redo_Original_1000"]
test = sorted(tmp.values(), reverse = True)
t = sum(test)
test = [j/t for j in test]
final["Redo_Original_1000"]['chem_data'] = test
final["Redo_Original_1000"]['color'] = 'Gray'


#i = "2000"
#tmp = chem_data[i]
#test = sorted(tmp.values(), reverse = True)
#test = [j/test[0] for j in test]
#final[i]['chem_data'] = test
#tmp = eng_data[i]
#test = sorted(tmp.values(), reverse = True)
#test = [j/test[0] for j in test]
#final[i]['eng_data'] = test
#final[i]['color'] = 'Black'

#i = "2000_2"
#tmp = eng_data[i]
#test = sorted(tmp.values(), reverse = True)
#test = [j/test[0] for j in test]
#final[i]['eng_data'] = test
#final[i]['color'] = 'Black'

#i = "2000_3"
#tmp = eng_data[i]
#test = sorted(tmp.values(), reverse = True)
#test = [j/test[0] for j in test]
#final[i]['eng_data'] = test
#final[i]['color'] = 'Black'

#for i in data:
#    tmp = data[i]
#    test = sorted(tmp.values(), reverse = True)
#    test = [j/test[0] for j in test]
#    final[i] = test
```
