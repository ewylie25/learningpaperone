#Market analysis exercises.
first of all I've pulled the last fifteen years of stock infos from yahoo.finance for a random company, Agilent. The choice is because it somehow sounded familiar. First of all we parse the csv, and we create a proper data structure, i.e. a list of dicts with all the interesting fields. Let's have the code speak by himself.

```python
from datetime import date 
with open('./agilent.csv') as f: # from yahoo.finance
    data=[]
    for x in f.readlines()[1:]: # we skip the first line as it contains the headers
        try:
            x=x.strip().split(',') # convert to a list
            d=date(*map(int,x[0].split('-')))# first entry is the time stamp, we convert to a date
            r=map(float,x[1:])# all the rest are floats, so we parse them
        except Exception as e:
            print e.message
            print "something went wrong when parsing {0}".format(x)
            continue
        data.append( {
                         'date':d,
                         'open':r[0],
                         'high':r[1],
                         'low':r[2],
                         'close':r[3],
                         'volume':r[4],
                         'close':r[5]
                         })
            
```

```python
data[-1]
```

```python
import vincent
vincent.core.initialize_notebook()
from math import log
revdata=data[::-1]#cronologically - [::-1]is the python-esque reverse
day2day=[]
dailyvolume=[]
for i,y in enumerate(revdata[1:]):
    day2day.append(log(revdata[i]['close'] / revdata[i-1]['close'])) #log(closing(i)/closing(i-1))
    dailyvolume.append(log(revdata[i]['volume'] / revdata[i-1]['volume'])) #log(closing(i)/closing(i-1))
#s=vincent.Scatter([x['close'] for x in revdata])
#s=vincent.Scatter([x for x in day2day])
s=vincent.Line({
                'index':range(len(day2day)),
                'Closure':[x for x in day2day],
                ' Volume':[x for x in dailyvolume] #the space is required since
                }, iter_idx='index') # the plots are stacked on z alphabetically. very annoying.
# s.scales['x'].type='time'
s.axis_titles(x='day',y='volume and closing value as log(t/t-1)')
s.legend(title='observables')
s.display()
```
